```python
import pandas as pd
```

```python
# Now, let's load the KEGG data from the Excel file
kegg_data = pd.read_excel('Data/KEGG_2021_Main.xlsx')

# Convert the KEGG DataFrame into a dictionary where each key is a column name
kegg_dict = {col: kegg_data[col].tolist() for col in kegg_data.columns}

# Now you have a dictionary 'kegg_dict' where each key corresponds to a column name, and the values are lists containing the entire columns from the KEGG DataFrame.
```

```python
# Specify the path to your CSV file
file_path = "Data/BDMC_Upload/BDMC_Final_Alzheimers Disease_D000544.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Select only the desired columns
selected_columns = df[['geneName', 'Frequency', 'BDMC_Score']]

# Create a new DataFrame with the selected columns
df = pd.DataFrame(selected_columns)

# Filter rows based on conditions
df = df[(df['Frequency'] >= 0) & (df['BDMC_Score'] >= 4)]

# Now, new_df contains only the 'geneName', 'Frequency', and 'BDMC_Score' columns from the original DataFrame.

# Add empty columns to the DataFrame df
for key in kegg_dict.keys():
    df[key] = None
```

```python
df
```

```python
# Create a designated DataFrame to store the results
result_df = df.copy()

# Initialize columns in the result DataFrame
for key in kegg_dict.keys():
    result_df[key] = 0

# Iterate through each 'geneName' value in the DataFrame
for index, gene_name in enumerate(result_df['geneName']):
    for key, value_list in kegg_dict.items():
        if gene_name in value_list:
            result_df.at[index, key] = 1

# Now, result_df contains the original data along with additional columns for each key in kegg_dict, with '1' for matches and '0' for non-matches.
```

```python
result_df
```

```python
# Calculate the sum of all values in each column (excluding 'geneName' column)
sum_values = result_df.drop(columns='geneName').sum()

# Create a new DataFrame for the sum row
sum_row = pd.DataFrame(data=sum_values).T
sum_row['geneName'] = 'Sum of Values'

# Reorder the columns to have 'geneName' as the first column
sum_row = sum_row[['geneName'] + [col for col in result_df.columns if col != 'geneName']]

# Concatenate the sum row with the result_df
result_df = pd.concat([sum_row, result_df], ignore_index=True)

# Now, result_df contains the sum row at the top with the sum of all values in each column, including the 'geneName' column.
```
