```python
import pandas as pd

# Define the input file path and the output file path
input_file_path = 'Data\KEGG_2021_Main - Copy.xlsx'
output_file_path = 'fa.xlsx'

# Load the input Excel file
input_df = pd.read_excel(input_file_path)

# Create a new Excel writer object
with pd.ExcelWriter(output_file_path, engine='openpyxl') as writer:

    # Iterate over columns from the input file
    for column in input_df.columns:
        # Create a new sheet in the output file with the same name as the column
        output_sheet_name = column
        input_column = input_df[column]

        # Create a DataFrame with a "Gene" column
        output_df = pd.DataFrame({'Gene': input_column})

        # Write the DataFrame to the output Excel sheet
        output_df.to_excel(writer, sheet_name=output_sheet_name, index=False)

print(f"Data manipulation and file creation completed. The output file '{output_file_path}' has been created.")
```

```python
import pandas as pd

# Load the "fa.xlsx" file
fa_df = pd.read_excel('fa.xlsx', sheet_name='Pathway 1')

# Load the "Comprehensive\combined_results.xlsx" file
combined_df = pd.read_excel('Comprehensive\combined_results.xlsx', sheet_name=None)

# Define the range of Pathways from 1 to 320
pathways = [f'Pathway {i}' for i in range(1, 321)]

# Iterate through each sheet in "combined_results.xlsx"
for sheet_name, sheet_data in combined_df.items():
    if all(pathway in sheet_data.columns for pathway in pathways):
        # Filter rows in the current sheet where all Pathway columns have a value of 1
        pathway_rows = sheet_data[sheet_data[pathways].eq(1).all(axis=1)]

        # Create a dictionary to map "geneName" values to "BDMC_Score" values
        gene_to_score = dict(zip(pathway_rows['geneName'], pathway_rows['BDMC_Score']))

        # Create new columns with the same names as the Pathway columns
        for pathway in pathways:
            fa_df[pathway] = fa_df['Gene'].map(gene_to_score)

# Save the updated "fa.xlsx" file
fa_df.to_excel('fa.xlsx', sheet_name='Pathway 1', index=False)
```

```python
import pandas as pd

# Load the "fa.xlsx" file
fa_df = pd.read_excel('fa.xlsx', sheet_name=None)

# Load the "Comprehensive\combined_results.xlsx" file
combined_df = pd.read_excel('Comprehensive\combined_results.xlsx', sheet_name=None)

# Create a new ExcelWriter object to save the updated "fa.xlsx" file
with pd.ExcelWriter('fa_updated.xlsx', engine='openpyxl') as writer:
    for isheet_name, isheet_data in fa_df.items():
        # Iterate through each sheet in "combined_results.xlsx"
        for sheet_name, sheet_data in combined_df.items():
            # Filter rows in the current sheet where the "Pathway 1" column has a value of 1
            pathway1_rows = sheet_data[sheet_data[isheet_name] == 1]

            # Create a dictionary to map "geneName" values to "BDMC_Score" values
            gene_to_score = dict(zip(pathway1_rows['geneName'], pathway1_rows['BDMC_Score']))

            # Create a new column with the same name as the current sheet in "combined_results.xlsx"
            isheet_data[sheet_name] = isheet_data['Gene'].map(gene_to_score)

        # Save the updated "fa_updated.xlsx" file with the current sheet
        isheet_data.to_excel(writer, sheet_name=isheet_name, index=False)

# Save the ExcelWriter to finalize the file
writer.close()
```

```python
import pandas as pd

# Load the "fa_updated.xlsx" file
fa_updated_df = pd.read_excel('fa_updated.xlsx', sheet_name=None)

# Create a new ExcelWriter object to save the modified file
with pd.ExcelWriter('fa_updated_modified.xlsx', engine='openpyxl') as writer:
    for sheet_name, sheet_data in fa_updated_df.items():
        # Fill empty cells with 0
        sheet_data = sheet_data.fillna(0)

        # Remove rows with 0 in the "Gene" column
        sheet_data = sheet_data[sheet_data['Gene'] != 0]

        # Save the modified sheet to the new file
        sheet_data.to_excel(writer, sheet_name=sheet_name, index=False)

# Save the ExcelWriter to finalize the modified file
writer.close()
```

```python
import pandas as pd

# Load the Excel file
file_path = "fa_updated_modified.xlsx"
xls = pd.ExcelFile(file_path)

# Create a writer to save the modified data to a new Excel file
output_file = "fa_updated_modified_with_counts.xlsx"
writer = pd.ExcelWriter(output_file, engine='openpyxl')

# Iterate through each sheet in the Excel file
for sheet_name in xls.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # Add a count row at the bottom of the sheet
    count_row = df.apply(lambda x: (x != 0).sum())
    count_row.name = 'Count'
    df = pd.concat([df, count_row.to_frame().T], axis=0)

    # Add a count column at the rightmost end of the sheet
    count_column = (df != 0).sum(axis=1)
    df['Count'] = count_column

    # Save the modified sheet to the new Excel file
    df.to_excel(writer, sheet_name=sheet_name, index=False)

# Save the new Excel file with the counts
writer.close()
```
