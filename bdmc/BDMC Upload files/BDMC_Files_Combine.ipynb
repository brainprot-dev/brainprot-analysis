```python
import pandas as pd
import os
import re


dir = '/content/drive/MyDrive/BrainProt_Version_3.0/2.BDMC/Data/BDMC_Upload'

name_append_1 = re.split("_",  os.listdir(dir)[1])[2] + "_"+ re.split("_",  os.listdir(dir)[1])[3][:-4]

final_df = pd.DataFrame(columns=['geneName', 'BDMC_Score'+ "_" + name_append_1, 'Frequency'+ "_" + name_append_1, 'Marker_Score'+ "_" + name_append_1])

for file in os.listdir(dir):
  print(file)
  if file.endswith(".csv"):
    tempdf = pd.read_csv(dir+ '/' + file)
    name_append = re.split("_", file)[2] + "_"+ re.split("_", file)[3][:-4]
    tempdf['Marker_Score'+ "_"+ name_append] = tempdf[['CTD_Marker', 'eDGAR_Marker', 'BIONDA_Marker', 'Dis_2_Marker']].sum(axis=1)
    tempdf.rename(columns = {'BDMC_Score':('BDMC_Score'+ "_" + name_append), 'Frequency': 'Frequency'+ "_" + name_append}, inplace = True)
    final_df = final_df.merge(tempdf[['geneName','BDMC_Score'+ "_" + name_append, 'Frequency'+ "_" + name_append, 'Marker_Score'+ "_" + name_append]],
                              on='geneName', how='outer')
```

```python
final_df.to_csv(dir+'/Merged_all_file.csv', index=False)
```

```python
import pandas as pd
import os
import re
import numpy as np
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

dir = '/content/drive/MyDrive/BrainProt_Version_3.0/5. BDPM/BDPM_FC_PVal'
name_append_1 = re.split("_",  os.listdir(dir)[0])[0]


final_df = pd.DataFrame(columns=['Uniprot', 'Fold Change'+ "_" + name_append_1, 't_test_p_value'+ "_" + name_append_1])

for file in os.listdir(dir):
  if file.endswith(".csv"):
    tempdf = pd.read_csv(dir+ '/' + file)
    name_append = re.split("_", file)[0]
    print(name_append)
    tempdf['Fold Change'] = MinMaxScaler().fit_transform(np.array(tempdf['Fold Change']).reshape(-1,1))
    tempdf.rename(columns = {'Fold Change':('Fold Change'+ "_" + name_append), 't_test_p_value': 't_test_p_value'+ "_" + name_append}, inplace = True)
    final_df = final_df.merge(tempdf[['Uniprot', 'Fold Change'+ "_" + name_append, 't_test_p_value'+ "_" + name_append]],
                              on='Uniprot', how='outer')
```

```python
final_df.to_csv(dir+'/BDPM_Merged_FC_pval_file.csv', index=False)
```
