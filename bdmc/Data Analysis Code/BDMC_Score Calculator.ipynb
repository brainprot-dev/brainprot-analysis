```python
BDMC_df = pd.read_csv("Creutzfeldt _Jakob_ Disease_D007562_BDMC_Test.csv")
Disease_Name = 'Creutzfeldt _Jakob_ Disease_D007562_BDMC_Test'
BDMC_df.head()
```

Calculation of Individul score: 'DP_Score', 'DC_Score', 'HP_Score', 'HC_Score', 'PC_Score'

```python
# create a list of column names to apply the formulas on
columns_to_apply_formulas_on = ['DISGENET_Score', 'Harmonizome_Score', 'Pubpular_Score', 'CTD_Score']

# calculate the new columns for all possible pairs of columns in the list using the formulas
for i in range(len(columns_to_apply_formulas_on)):
  if columns_to_apply_formulas_on[i] not in BDMC_df.columns:
    BDMC_df[columns_to_apply_formulas_on[i]] = 0
    continue
  for j in range(i+1, len(columns_to_apply_formulas_on)):
    if columns_to_apply_formulas_on[j] not in BDMC_df.columns:
      BDMC_df[columns_to_apply_formulas_on[j]] = 0
      continue
    # create a new column name by concatenating the two column names with "top_" in the beginning
    colname = 'top_' + columns_to_apply_formulas_on[i].lower() + '_' + columns_to_apply_formulas_on[j].lower()
    # apply the formula to calculate the value for each row in the new column
    BDMC_df[colname] = ((BDMC_df[[columns_to_apply_formulas_on[i], columns_to_apply_formulas_on[j]]].mean(axis=1)) +
                                (1 - (BDMC_df[[columns_to_apply_formulas_on[i], columns_to_apply_formulas_on[j]]].std(axis=1) /
                                        BDMC_df[[columns_to_apply_formulas_on[i], columns_to_apply_formulas_on[j]]].mean(axis=1))) / 2)

#dis2_df = pd.read_csv("/filepath")
#dis2_names = list(dis2_df['Name'])

#BDMC_df['Dis_2_Marker'] = BDMC_df['Gene Name'].apply(lambda x: 1 if x in dis2_names else 0)

BDMC_df = BDMC_df.loc[:, ['SwissProt_Accessions', 'geneName', 'DISGENET_Score', 'Harmonizome_Score', 'Pubpular_Score', 'CTD_Score',
                          'top_disgenet_score_harmonizome_score', 'top_disgenet_score_pubpular_score', 'top_disgenet_score_ctd_score',
                          'top_harmonizome_score_pubpular_score', 'top_harmonizome_score_ctd_score', 'top_pubpular_score_ctd_score']]


BDMC_Final_df = BDMC_df.rename(columns={
    'top_disgenet_score_harmonizome_score': 'DH_Score',
    'top_disgenet_score_pubpular_score': 'DP_Score',
    'top_disgenet_score_ctd_score': 'DC_Score',
    'top_harmonizome_score_pubpular_score': 'HP_Score',
    'top_harmonizome_score_ctd_score': 'HC_Score',
    'top_pubpular_score_ctd_score': 'PC_Score'
    })

BDMC_Final_df.head()


```
